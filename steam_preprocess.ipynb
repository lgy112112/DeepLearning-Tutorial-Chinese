我们常用的工具首先是`pandas`

`pandas`是一个强大的数据分析和处理工具库，广泛应用于数据科学和机器学习领域。它提供了高效、易用的数据结构和数据分析功能，使得数据处理任务更加简洁和高效。以下是`pandas`在数据分析中的几个关键作用：

4. **数据操作与变换**：`pandas`提供了灵活的数据操作方法，如筛选、排序、合并、连接、透视表等，帮助用户对数据进行各种复杂变换。
5. **数据可视化**：`pandas`可以与`matplotlib`、`seaborn`等可视化库结合使用，轻松创建各种图表，以直观地展示数据特征。

# 图形
是最直观的观察方式，我们可以用matplotlib和seaborn库来绘制图形，用于解释我们的数据。接下来，我们通过直方分布图来了解，每一个特征的分布情况。

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 加载数据
train_path = 'steam_train.csv' #我在这里使用了相对路径格式
test_path = 'steam_test.csv'

train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

# 绘制直方图，叠加显示
plt.figure(figsize=(15, 20))
for i, col in enumerate(train_data.columns):
    plt.subplot(8, 5, i + 1)
    sns.histplot(train_data[col], color="blue", label='Train', kde=True, stat="density", linewidth=0, alpha=0.6)
    if col in test_data.columns:  # 测试集不包含目标变量
        sns.histplot(test_data[col], color="red", label='Test', kde=True, stat="density", linewidth=0, alpha=0.6)
    plt.title(col)
    plt.legend()
    plt.tight_layout()

plt.show()
````

## 我们可以看到训练集和测试集中特征V1~V37的分布，
他们大多都是像正态分布的样子（很多其实不是，但这不是我们要讨论的范围）。

其中，我们可以看到几个**比较丑陋**的分布，比如*v5 v9 v11 v17 v19  v20 v21 v22 v24 v28*。仔细观察他们，为何与众不同呢？

因为他们要么就是错峰，要么就是分布泛散，总之可以粗略归结为**训练集中和测试集中分布不一致**。我们靠直觉来想，**如果我给你一份线性代数的学习资料（训练集数据），要你去考大学物理（进行测试集预测），虽然两者有点关系，但是还是很不同，这会导致你大物考得很差（测试集效果差）**

既然我们考的是大物，那我们学习的资料肯定就要是大物，那我们就要对我们的学习资料进行调整，来**对训练集进行预处理**，使得两者分布一致。

预处理（特征工程）的方法有很多，但总结起来无非就是：**将有用的特征留下并且突出（也可以制造有用的特征），将无用的特征看作噪声抑制或者删除**。我们在此做的就是删除没用的特征。

## 或许
你担心，删除的特征怎么就没用呢？你当然可以保留并且给模型进行训练，但减少数据减少特征可以让训练更快、让模型专心学习更有用的特征，有时候你要做出取舍。

## 那么如何判断特征是否有用？我们可以制作`heatmap`相关系数热图来观察👇

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 加载数据
# 定义训练数据文件的路径
train_path = 'steam_train.csv'

# 使用pandas的read_csv方法读取CSV文件，并将数据加载到DataFrame对象中
train_data = pd.read_csv(train_path)

# 计算每个特征与目标变量之间的相关系数
# 使用DataFrame的corr方法计算各列之间的相关系数矩阵
# 然后提取与目标变量'target'的相关系数，并移除'target'自身的相关系数
correlation_with_target = train_data.corr()['target'].drop('target')

# 按绝对值降序排列相关系数
# 为了找到最相关的特征，将相关系数取绝对值并按降序排列
correlation_with_target = correlation_with_target.abs().sort_values(ascending=False)

# 绘制热图
# 设置图形的尺寸（宽度10，高度8）
plt.figure(figsize=(10, 8))

# 使用seaborn的heatmap方法绘制热图
# to_frame()方法将Series对象转换为DataFrame，annot=True表示在热图中显示数值
# cmap='coolwarm'表示使用coolwarm配色方案，cbar=False表示不显示颜色条
sns.heatmap(correlation_with_target.to_frame(), annot=True, cmap='coolwarm', cbar=False)

# 设置热图的标题
plt.title('Correlation of Features with Target')

# 显示热图
plt.show()
````

我们可以看到，暖色调的特征与target值关系密切；而其他特征就不那么密切。通过比较，我们要删除的特征都在冷色调列中，佐证了我们做法的正确性。

````python
import pandas as pd

# 加载数据
train_path = 'steam_train.csv'
test_path = 'steam_test.csv'

train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

# 定义要删除的列
columns_to_drop = ['V5', 'V9', 'V11', 'V17', 'V19', 'V20', 'V21', 'V22', 'V24', 'V28']

# 从训练集和测试集中删除这些列
train_data_dropped = train_data.drop(columns=columns_to_drop)
test_data_dropped = test_data.drop(columns=columns_to_drop)

# 定义新的CSV文件路径
new_train_path = 'steam_train_new.csv'
new_test_path = 'steam_test_new.csv'

# 将修改后的数据保存为新的CSV文件
train_data_dropped.to_csv(new_train_path, index=False)
test_data_dropped.to_csv(new_test_path, index=False)

print("已成功删除指定列并保存新的CSV文件。")
````

# 最后
我们可以进行归一化

````python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# 加载数据
train_path = 'steam_train_new.csv'
test_path = 'steam_test_new.csv'

train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

# 创建MinMaxScaler对象
scaler = MinMaxScaler()

# 对训练集进行归一化
# 仅对以'V'开头的特征列进行归一化，不包括目标变量
train_features = train_data.filter(regex='^V')
train_target = train_data['target']
train_features_scaled = scaler.fit_transform(train_features)

# 对测试集进行归一化
# 使用与训练集相同的scaler进行转换，确保两者在同一尺度下
test_features = test_data.filter(regex='^V')
test_features_scaled = scaler.transform(test_features)

# 将归一化后的特征和目标变量重新组合成DataFrame
train_data_scaled = pd.DataFrame(train_features_scaled, columns=train_features.columns)
train_data_scaled['target'] = train_target.values

test_data_scaled = pd.DataFrame(test_features_scaled, columns=test_features.columns)

# 保存归一化后的数据
train_scaled_path = 'steam_train_scaled.csv'
test_scaled_path = 'steam_test_scaled.csv'
train_data_scaled.to_csv(train_scaled_path, index=False)
test_data_scaled.to_csv(test_scaled_path, index=False)

print("训练集和测试集已归一化并保存。")
````

为了确保训练集和测试集的数据在同一尺度下进行归一化，我们需要使用在训练集上拟合的MinMaxScaler对象对测试集进行转换。这可以保证训练集和测试集使用相同的缩放参数（最小值和最大值）进行归一化，从而使得两者在同一尺度下。

### 详细步骤解释：

1. **创建并拟合Scaler**：
   - 在训练集上创建并拟合MinMaxScaler。这一步会计算训练集中每个特征的最小值和最大值，并存储这些值。

2. **转换训练集**：
   - 使用拟合好的Scaler对训练集进行转换。这会将训练集的每个特征值缩放到0到1之间。

3. **转换测试集**：
   - 使用同一个Scaler对测试集进行转换。因为这个Scaler已经存储了训练集的最小值和最大值，所以在转换测试集时，会使用相同的缩放参数，确保两者在同一尺度下。



### 确保同一尺度的关键步骤：

1. **创建并拟合Scaler**：
    ```python
    scaler = MinMaxScaler()
    train_features_scaled = scaler.fit_transform(train_features)
    ```
    - 这里，我们在训练集上拟合Scaler。`fit_transform`方法同时进行拟合和转换。`fit`步骤计算并存储训练集的最小值和最大值。

2. **转换测试集**：
    ```python
    test_features_scaled = scaler.transform(test_features)
    ```
    - 使用已经在训练集上拟合好的Scaler对测试集进行转换。`transform`方法使用已经存储的训练集的最小值和最大值对测试集进行缩放。

通过这些步骤，我们确保了训练集和测试集使用相同的缩放参数进行归一化，从而使得两者在同一尺度下。这样可以防止因为不同的缩放参数导致的训练和测试数据分布不一致问题，提高模型的泛化能力和性能。

有很多标准化/归一化的方法，这里只介绍一种。你可以去探索其他，比如zscore，勤加搜索，看别人怎么写的然后缝合过来是很重要的学习。
